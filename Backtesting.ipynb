```python
import vectorbt as vbt
import yfinance as yf
from datetime import datetime, timedelta
import talib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1. BTC example

```python
end_time = datetime.now()
start_time = end_time - timedelta(days=20)
```

```python
btc = yf.download(
    'BTC-USD', 
    start=start_time, 
    end=end_time,
    interval='1h')
btc
```

```python
hammer = talib.CDLHAMMER(btc[('Open', 'BTC-USD')], btc[('High', 'BTC-USD')], btc[('Low', 'BTC-USD')], btc[('Close', 'BTC-USD')])
hammer
```

```python
hanging_man = talib.CDLHANGINGMAN(btc[('Open', 'BTC-USD')], btc[('High', 'BTC-USD')], btc[('Low', 'BTC-USD')], btc[('Close', 'BTC-USD')])
hanging_man
```

```python
buys=(hammer == 100)
sells=(hanging_man == -100)
pf_btc = vbt.Portfolio.from_signals(btc[('Close', 'BTC-USD')], entries=buys, exits=sells, fees=0.005)
pf_btc.stats()
```

```python
pf_btc.plot().show()
```
